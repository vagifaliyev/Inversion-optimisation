The forced one-dimensional viscous Burgers' equation occurs frequently in fluid dynamics and acoustics. In form it is very similar to the Navier-Stokes equations you saw in ACSE-3,

$$ \frac{\partial u}{\partial t} +u\frac{\partial u}{\partial z} = \nu \frac{\partial^2 u}{\partial z^2} +r(z).$$

Here $u$ is the variable of interest, $z$ is the independent variable representing space in the domain and $r(z)$ is a given forcing term. For numerical convenience it is also common to write the advective term $u\frac{\partial u}{\partial z}$ in a conservative form,
$$ \frac{\partial u}{\partial t} +\frac{1}{2}\frac{\partial u^2}{\partial z} = \nu \frac{\partial^2 u}{\partial z^2} +r(z).$$

You have been provided with observational data of $u$ for a number of time levels at $z=5$ and $z=-5$ from a Burger's model system in a periodic domain of period 20. In the region $[-10, 10]$ $r(z)$ takes the form $100\sin\left(\frac{3t}{8}\right)\exp(-100z^2)$.

* 1.  A single 4D-Var assimilation:
   
4D-Var is the time-aware equivalent of the 3D-Var method. Although, it works under the assumption that the model is perfect so problems can be arise if model errors are large. 
   
Implicitly improves $B$ over time window, uses all information in time window and is cheaper than equivalent Kalman filter.
   
However, new $B$ not saved, backing out individual error bounds significant extra work and is costlier that OI/3D-Var.
   
The multiple method is a technique designed to reduce the cost of solving the variational problem, e.g. by reducing the resolution of the increments.

With 4D-Var it is usually not affordable to solve the variational problem at the full model resolution due to high cost. Most of the complexity of the analysis lies in the synoptic scales, due most background errors residing there. Thus, if the increments are right at the synoptic scales, then we can assume that the smaller scales to be close to realistic features by the model dynamics.

To be concise, to completely neglect the small scales in the analysis procedure is not reccomended since it is important in the comparison between the observations with the background state. In other words, it is much better for accuracy to solve in single if the cost is not a major issue to be worried about. 

* 2.  Multiple 4D-Var assimilation events:

This method is much better than 3D-var cause it uses observations at the correct time, calculates analysis at the correct time 
and implicitly generates ï¬‚ow-dependent forecast error covariance. 

A special property of the 4D-Var analysis is that in the middle of the time interval it uses all the observations simultaneously, not just the ones before the analysis time. It can be said that 4D-Var is a smoothing algorithm. 

* 3.  Multiple iterations of the ordinary Kalman filter:
  
Works with known error bounds and has the most appropriate B.
  
However, calculating new $B$ is really costly and information only carried forwards, not back.
  
Over a given time interval, under the assumption that the model is perfect, with the same input data , the 4D-Var analysis at the end of the time interval is equal to the Kalman filter analysis at the same time.
 

* 4.  Multiple 3D-Var assimialtion events:
  
Only uses single $\mathbf{B}$ and allows nonlinear observations.
  
However, there is no time dependence and iteration to nonlinear solution are more costly.

* 5.  Multiple OI assimilation events:
  
Var is usually better than OI due to it being able to handle lots of observations ,copes better with nonlinear observation operator H and solves for the whole domain at once.
  
OI solves the problem separately in a set of subspaces, which can be defined by the observation selection, however the result is not as optimal as a global least-squares analysis.
  
* 6.  A single OI assimilation:
  
The advantage of OI is its simplicity of implementation and its relatively small cost if the right assumptions can be
made on the observation selection.
    
However, spurious noise is produced in the analysis fields because different sets of observations
and background error models are used on different parts of the model state.

You may assume that the observations are independent, with Gaussian errors with standard deviation 0.05 (at $z=-5$ ) and 0.1 (at $z=5$). The supplied observational data and background error covariance matrix can be loaded using the following code block:

```python
# observations data is in array form y[timelevel, location]
# where location 0 is z=-5 and location 1 is z=5
#
# The actual time value is stored in the t_obs array

import numpy as np
data = np.load('burgers_observations.npz')
t_obs = data['t_obs']
y = data['y']

# B is the (sparse) background error covariance matrix
def generate_B_matrix(z):
    B = np.empty([len(z), len(z)])
    
    for _ in range(len(z)):
        B[_, :] = 0.2**2*np.exp(-4*(z[_]-z)**2)*((z[_]-z)**2<=4)
    return B
```

- Implement your best data assimilation strategy to provide a forecast at $t=100$ for an initial background at $t=0$ of $u=1$ (i.e $x_b$=1) using the observations, background error covariance matrix and observation statistices provided. 
- Write a short (i.e three paragraph or less) description of your method, and of any potential issues that it will not be able to cope with.

For your convenience, you may use the following forecast model, or write your own.


```python
# %matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import inv

nz = 61
z = np.linspace(-10, 10, nz)
dz = z[1]-z[0]
u = np.full(z.shape, 1.0)
x_b = np.full(z.shape, 1.0)
nu = 0.4
dt = 0.05
#background error covariance matrix
B = generate_B_matrix(z)

## Use forward Euler for the (non-linear) advection  and Crank-Nicolson for diffusion
def rhs(U, dt, t):
    out = np.empty(U.shape)
    out[0] = U[0]-0.5*dt/dz*((U[1]**2-U[-1]**2)-nu*(U[-1]-2*U[0]+U[1])/dz)
    out[1:-1] = U[1:-1]-0.5*dt/dz*((U[2:]**2-U[:-2]**2)-nu*(U[:-2]-2*U[1:-1]+U[2:])/dz)
    out[-1] = U[-1]-0.5*dt/dz*((U[0]**2-U[-2]**2)-nu*(U[-2]-2*U[-1]+U[0])/dz)
    return out+dt*10*np.sin(.375*np.pi*(t+0.5*dt))*np.exp(-100*z**2)

def lhs_matrix(u, dt):
    mat = np.eye(u.shape[0])
    mat[0,-1] -= 0.5*dt*nu/dz**2
    mat[0, 0] += dt*nu/dz**2
    mat[0, 1] -= 0.5*dt*nu/dz**2
    for _ in range(1, u.shape[0]-1):
        mat[_, _-1] -= 0.5*dt*nu/dz**2
        mat[_, _] += dt*nu/dz**2
        mat[_, _+1] -= 0.5*dt*nu/dz**2
    mat[-1,-2] -= 0.5*dt*nu/dz**2
    mat[-1, -1] += dt*nu/dz**2
    mat[-1, 0] -= 0.5*dt*nu/dz**2
    return mat

def burgers_equation(u0, dt, t_final):
    t = 0.0
    u = u0
    out = [u0]
    mat = lhs_matrix(u0, dt)
    matinv = inv(mat)
    while t<t_final:
        for _ in range(10):
            u = matinv.dot(rhs(u, dt, t))
            t += dt
        out.append(u)
    return out

data = burgers_equation(u, dt, 100)

plt.plot(z, data[-1]);
```

## Code written 

```python


R = [[0.05**2, 0],
     [0, 0.1**2]]

# -- generate h --
# observations to fall exactly on model timelevels you 
# so this section is made much simpler 
# else might need extra tweaking
h = np.zeros((2, nz))
for _ in range(nz):
    if z[_] == -5:
        h[0,_] = 1
    elif z[_] == 5:
        h[1,_] = 1

# updated every time-step
def generate_M_matrix(x):
    M = np.zeros((nz, nz))
    for _ in range(nz):
        #diagonal
        M[_][_] = 1+dt*(nu/(dz**2))
        #upper diag
        if _ < nz-1:
            M[_][_+1] = dt*(-x[_]/(dz)-nu/(2*dz**2))
        else: 
            M[_][0] = dt*(-x[_]/(dz)-nu/(2*dz**2))
        #lower diag
        if _ == 0:
            M[0][nz-1] = dt*(x[_]/(dz)-nu/(2*dz**2))
        else:
            M[_][_-1] = dt*(x[_]/(dz)-nu/(2*dz**2))
    return M

### Calculate the adjoint model under the tangent linear assumption
### gives lamba_0
def adjoint_model(x, y, h):
    # calculating lambda_N (the lambda in the final step)
    # needs R
    H = y[-1]- h@x[-1]
    l = -h.T@R@H
    obv_id = len(y)-1
    #solve backwards for the other lambdas
    for _ in range(int((len(y)-1)/(dt*10)),0,-1):
    # copy lambda into the time step before
        tmp = l.copy()
        
        tmp[1:-1] = l[1:-1] *(1- dt*nu /dz**2) + l[:-2]*((dt*nu)/(2*dz**2)) + l[2:]*((dt*nu)/(2*dz**2))
        tmp[0] = l[0] * (1-(dt*nu)/dz**2) + l[-1]*((dt*nu)/(2*dz**2)) + l[1]*((dt*nu)/(2*dz**2))
        tmp[-1] = l[-1] * (1-(dt*nu)/dz**2) + l[-2]*((dt*nu)/(2*dz**2)) + l[0]*((dt*nu)/(2*dz**2))
        
        M = generate_M_matrix(x[_])
        
        # calculate lambda_i; tmp == M.T * lambda
        l = inv(M.T)@tmp
        if _*(dt*10) in t_obs:
            obv_id -= 1
            l -= h.T@R@(y[obv_id]-h@x[_])
    return l

def j(x0, y, h):
    x = burgers_equation(x0, dt, 100)
    j_b0 = 0.5*(x0-x_b).dot(B.dot((x0-x_b)))
    j_b = j_b0
    for _ in range(len(y)):
        x_id = int(_/(dt*10))
        j_b -= 0.5*((y[_]-h@x[x_id]).T@R@(y[_]-h@x[x_id]))
    return j_b

### the function giving the gradient with respect to x_0
def jac(x0, y, h):
    #solving the forward model
    x = burgers_equation(x0, dt, 100)
    #calculating the gradient 
    jac= B.dot(x0-x_b) + adjoint_model(x, y, h)
    return jac

from scipy.optimize.linesearch import line_search_armijo as line_search
### iterate with a basic linesearch to minimise the cost function and generate the 
### analysis state.
x_a = np.ones(len(x_b))
#print(x_a)
j0 = j(x_a, y, h)
#print("j0 = ",j0)
for _ in range(100):
    # search direction as minus jacobian
    pk = -jac(x_a, y, h)
    # 
    res = line_search(j, x_a, pk, -pk, j0, args=(y, h), alpha0=0.5)
    x_a += res[0]*pk 
    j0 = res[2]
    print(res)
    #stopping criterion 
    if res[0]<1e-5:
        print("-- broke --")
        break
        
# ------------------------------------------------------------------------------------
```

```python
## lets do some graphing 
x = range(80)

#lets extract -5 and 5 values
u_a1 = []
u_a2 = []

# adding the corresponding values for graphing 
rs = burgers_equation(x_a, dt, 100)
t_step = len(rs)
for _ in range(nz):
    if z[_] == -5:
        for i in range(t_step):
            if (_*i*dt*10).is_integer():
                u_a1.append(rs[i][_])
    elif z[_] == 5:
        for i in range(t_step):
            if (_*i*dt*10).is_integer():
                u_a2.append(rs[i][_])

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

#for z = -5 
ax1.plot(x, y, c="blue", marker="o", label='obs')
ax1.plot(x, u_a1[:80], c="orange", marker="o", label='as')
ax1.set_title("$z = -5$", fontsize=16)
ax1.grid('on')

#for z = 5 
ax2.plot(x, y, c="blue", marker="o", label='obs')
ax2.plot(x, u_a2[:80], c="orange", marker="o", label='as')
ax2.set_title("$z = + 5$", fontsize=16)
ax2.grid('on')
plt.show()
```
