# Question A1

Consider a linear inversion problem written in the form 

$$G\boldsymbol{m} = \boldsymbol{d}.$$

The following four mathematical expressions give possible solutions to this problem


1. $\boldsymbol{m} =  G^+\boldsymbol{d},$


2. $\boldsymbol{m} = G^T(GG^T)^{-1}\boldsymbol{d},$


3. $\boldsymbol{m} = G^{-1}\boldsymbol{d},$


4. $\boldsymbol{m} = (G^T G)^{-1} G^T  \boldsymbol{d}.$


For each of these approaches explain in words what type of inversion problem the approach can be used to solve, what properties the corresponding solution will possess, and any potential issues that might arise when attempting to utilise this solution approach.

# Answer A1

## --- 1 --- 

**Mixed-determined problems** 

In the scenerio where  $G$  is  not  square  so  that  $G^{-1}$ does  not  exist,  and  both  $G^TG$ and $GG^T$ are singular matrices, then the generalised inverse $G^+$ can be used, also refered as the *pseudo-inverse or the Moore-Penrose inverse*.

$$G^+=G^{-1}$$

$G^+$ always exists for any non-zero matrix $G$ so that the generalised inverse solution is given by  

$$\boldsymbol{m}^+ = G^+\boldsymbol{d}$$

The generalised inverse is preferable in small scale problems, especially when the quality of the results wants to be analysed. 


## --- 2 --- 

**Under-determined problems**

Fewer equations than unknowns exist in this scenerio, thus $m<n$. Hence, the matrix $G^TG$ will be singular. However the $m \times m$ matrix $GG^T$ will not be singular if the equations are independent and consistent.  

Under such case following solution can be constructed:

$$\boldsymbol{m} = G^T(GG^T)^{-1}\boldsymbol{d}$$

where the matrix $G^T(GG^T)^{-1}$ can be used instead of inverse of $G$ since:

$$G\left( G^T (G G^T)^{-1} \right) = I$$

which is also refered to as *minimum-norm inverse*. This generally perfectsly fits the data, however this may not be the only exact solution. This is due to the problem having a null space, since there are fewer equations than unknowns, thus infinite number of solutions will satisfy the data equally well. 

## --- 3 --- 

**Equi-determined problems - square linear systems**

If $G$ is not singular, then equivalently the linear equations will all be independent, there will be only one solution which will be an exact fit to the data. Solution is thus given by:

$$\boldsymbol{m} = G^{-1}\,\boldsymbol{d}$$

However, the solution being exact is more of a mathematical statement rather than a physical one. This is due to the real life data always containing some kind of uncertainties which will inevitably influence the solution

In the case that the matrix $G$ is singular, then the inverse of $G$ will not exist, thus above equation will be invalid for finding a solution. 

In such scenerio, either the equations are independent resulting in infinitely many solutions or equations are incompatible thus no solution will exist which exactly fits the data.


## --- 4 --- 

**Over-determined problems** 

In this case, there are more independent equations than unknowns.  
The system:  

$$G\, \boldsymbol{m} = \boldsymbol{d}$$

where $G$ is an $m\times n$ matrix, also $m>n$.

The inverse of $G$ is *generally* not defined, and there is no solution for $m$ that will be satisfied fully.

However, another solution can be found by instead solving the related equation which is called **normal equation**:

$$G^T \,G\, \boldsymbol{m} = G^T \,\boldsymbol{d}$$

The solution to the normal equation, provided that the matrix is is not *singular*, also the $n\times n$ matrix $G^T \,G$ will be  square and symmetric thus : 
 
$$\boldsymbol{m} = (G^T \,G)^{-1}G^T \,\boldsymbol{d}$$

Through this approach, **least squares solution** will be generated for the problem.

Its inverse will exist provided that $G$ is of full column rank ($\text{rank}(G)=n$). In the scenerio where $G$ is not full rank, so that $G^TG$ is singular. Such problems are defined as mixed-determined.

# Question A2

Consider the rectangular matrix

$$G = 
\begin{pmatrix}
2 & 2 & 10 & 10 \\
2 & 3 & 14 & 12 \\
4 & 5 & 24 & 22 
\end{pmatrix}.
$$

Convert this matrix into reduced-row-echelon form. 

What is the rank of this matrix? What dimensions are its range and its null space?

Find the appropriate number of vectors which can be used as a basis to span the null space.

Now consider the data vector
$$\boldsymbol{d} = \begin{pmatrix}
26 \\
30\\
56
\end{pmatrix}.$$


Use the pseudo-inverse (you may make use of `np.linalg.pinv`) to find a solution to the problem $G\boldsymbol{m}=\boldsymbol{d}$.

Demonstrate though some examples that adding any multiples of the vectors spanning the null space to this solution (i.e. adding any vector from within the null space) also provides a valid solution to this problem.

Demonstrate that the solution obtained using the pseudo-inverse is the minimal norm solution to this problem.

# Answer A2

```python
G = Matrix([[2,2,10,10],[2,3,14,12],[4,5,24,22]])
G_rref = np.array([[1,0,1,3],[0,1,4,2],[0,0,0,0]])

print("RREF: \n", G_rref, "\n")

print("Rank: ", G.rank(), "\n")

print("Range: 2", "\n")

null = G.nullspace()
null1 = np.array([-1,-4,1,0])
null2 = np.array([-3,-2,0,1])

print("Null_space:\n", null1, "\n", null2)
```

```python
#solution to Gm=d using pseudo-inverse
G = np.array([[2,2,10,10],[2,3,14,12],[4,5,24,22]])
d = np.array([26,30,56])

G_plus = np.linalg.pinv(G) 
m = G_plus.dot(d) #G_plus@d
print("m is:", m)
```

```python
once = True

#trying range of alpha and beta values to test
for i in range(-10,10):
    for j in range(-10,10):
        #addition
        x = m + i*null1 + j*null2 
        #set x_prev to current values once at the start
        if once == True:
            x_prev = x
            once = False
        #checking if current answer agrees with the previous one
        #if not return error and the array of answers
        if np.allclose(G@x, G@x_prev) == False:
            print("non-valid solution detected")
            print(G@x)
            print(G@x_prev)
        #set current x values as previous values for the next operation
        x_prev = x

#display answer
x_m = G@x
print(x_m)
```

```python
from numpy import linalg as nla

# is x_m the minimum norm solution?
G_ri = G.T @ sl.inv(G@G.T)

x_m = G_ri@d

# plot the norm of the vectors we get by adding on multiples of the numm vector
fig = plt.figure(figsize=(5, 5))
ax1 = fig.add_subplot(111)

mult = np.linspace(-1000,1000,100)
norms = []
for mu in mult:
    norms.append(nla.norm(x_m + mu*null1 +mu*null2))

ax1.plot(mult,norms)

# as we hoped for the norm is at a minimum when the multiplier is zero!
```

# Answer A3

One of the reasons SVD was used for image compression is because the first components of these matrix can be used to give a close approximation of how the actual matrix looked like. 

If we assume that the most important information of $U$ and $V^T$ is stored on its first components (let's assume first column). Then the Matrix $A$ can be approximated as:

$$ A' = U_p\Sigma_p V_p^{T} 
     = [(mx1)(1x1)](1xn)
     = (mx1)(1xn)
     = (mxn)
$$

This computation will be the same size as the original matrix which I believe is a good approximation of the original matrix. For even a closer approximation, the next column of values can be used also. 

Another method is to use the fact that the sum of all singular values represents the visual compression while the amount of singular values represents the memory compression. 

Read Article: 

https://www.math.utah.edu/~goller/F15_M2270/BradyMathews_SVDImage.pdf

```python
# PIL = Python Imaging Library
from PIL import Image
img_colour = Image.open('./Imperial_College_Royal_School_of_Mines.jpg')
# convert to grey scale
img_grey = img_colour.convert('LA')
# what does LA mean? ... https://pillow.readthedocs.io/en/stable/handbook/concepts.html#concept-modes

fig = plt.figure(figsize=(8, 12))
ax1 = fig.add_subplot(111)
ax1.imshow(img_colour)
```

```python
# as we converted to grey scale, we can represent 
# the image via a matrix with entries representing 
# pixels and the corresponding value representing
# a shade of grey between black and white

# use getdata, first band - https://www.geeksforgeeks.org/python-pil-image-getdata/
img_mat = np.array(img_grey.getdata(band=0), float)
# turn into matrix of the correct shape
img_mat.shape = (img_grey.size[1], img_grey.size[0])

fig = plt.figure(figsize=(8, 12))
ax1 = fig.add_subplot(111)
# plot using a gray colormap
ax1.imshow(img_mat, cmap='gray')
```

```python
# find the SVD
U, sigma, VT = np.linalg.svd(img_mat)

# print('np.shape(U):', np.shape(U))
#print('np.shape(sigma):', np.shape(sigma))
# print('np.shape(VT):', np.shape(VT))

# n = 10
# print('\nFirst ' + str(n) +' singular values:', *list(sigma[:n]), sep='\n')
# print('\nLast '+ str(n) +' singular values:', *list(sigma[-n:]), sep='\n')

#for n in range (1,2):
#    frac = sum(sigma[:n]) / sum(sigma)
#    print('\n','Metric: ',frac)
```

```python
size = np.shape(sigma)
x = np.arange(0, size[0])
y = []
found = False

for n in range (0,size[0]):
    frac = sum(sigma[:n]) / sum(sigma)
    y.append(frac)
    if n == size[0]-1:
        print("Reached 1.0 at ", n)

fig, ax = plt.subplots(1)
ax.plot(x, y, 'k--', linewidth=1.5)
plt.hlines(y=1, xmin=0, xmax=size[0], colors='r', lw=2)
ax.grid(True)
plt.show()
```
